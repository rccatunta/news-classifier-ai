```python
import pandas as pd
import re, string, unicodedata
df = pd.read_csv('noticias.csv')
df.head()
```

```python
def remove_punctuation(words):
    """Remove punctuation from list of tokenized words"""
    new_words = []
    for word in words:
        new_word = re.sub(r'[^\w\s]', '', word)
        if new_word != '':
            new_words.append(new_word)
    return new_words
```

```python
def to_lowercase(words):
    """Convert all characters to lowercase from list of tokenized words"""
    new_words = []
    for word in words:
        new_word = word.lower()
        new_words.append(new_word)
    return new_words
```

```python
def remove_non_ascii(words):
    """Remove non-ASCII characters from list of tokenized words"""
    new_words = []
    for word in words:
        new_word = unicodedata.normalize('NFKD', word).encode('ascii', 'ignore').decode('utf-8', 'ignore')
        new_words.append(new_word)
    return new_words
```

```python
import nltk 
from nltk.tokenize import word_tokenize
titles = df.titulo.str.cat(sep= ' ')
tokens = word_tokenize(titles)
tokens = remove_non_ascii(tokens)
tokens = remove_punctuation(tokens)
tokens = to_lowercase(tokens)
vocabulary = set(tokens)
print(len(vocabulary))

frequency_dist = nltk.FreqDist(tokens)
sorted(frequency_dist,key=frequency_dist.__getitem__, reverse = True)[0:50]
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('spanish'))
tokens = [w for w in tokens if not w in stop_words]
frequency_dist = nltk.FreqDist(tokens)
sorted(frequency_dist,key=frequency_dist.__getitem__, reverse = True)[0:50]
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud().generate_from_frequencies(frequency_dist)

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()
df['categoria'] = LE.fit_transform(df['categoria'])
df.head()
```

```python
cnt = 0
for row in df['titulo']:
    try:
        title = df.loc[cnt,'titulo']
        tokens_title = word_tokenize(title)
        tokens_title = remove_non_ascii(tokens_title)
        tokens_title = remove_punctuation(tokens_title)
        tokens_title = to_lowercase(tokens_title)
        stop_words = set(stopwords.words('spanish'))
        tokens_tile = [w for w in tokens_title if not w in stop_words]
        new_title = ""
        for token in tokens_title:
            if token in tokens:
                new_title = new_title + token + " "
        df.loc[cnt,'titulo'] = new_title
        # print(new_title)
    except ValueError:
        pass
    cnt+=1
```

```python
import numpy as np
from sklearn.naive_bayes import MultinomialNB #is better than GaussianNB for words
from sklearn.feature_extraction.text import CountVectorizer
```

```python
X = df['titulo']
Y = df['categoria']
```

```python
count_vector = CountVectorizer()
count_vector.fit(X)
names = count_vector.get_feature_names()
doc_array = count_vector.transform(X).toarray()

frecuency_matrix = pd.DataFrame(data = doc_array, columns = names)
frecuency_matrix
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(doc_array,Y,test_size=0.1)
```

```python
#Model Training
model = MultinomialNB()
model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score

predicted = model.predict(X_test)

print(accuracy_score(Y_test,predicted))
```

```python
import seaborn as sn
confusion_matrix = pd.crosstab(Y_test, predicted, rownames=['Actual'], colnames=['Predicho'])
sn.heatmap(confusion_matrix, annot=True)
```
